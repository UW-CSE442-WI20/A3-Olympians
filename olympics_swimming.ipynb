```python
import pandas as pd
```

```python
olympics = pd.read_csv('../athlete_events.csv')
```

```python
olympics.head()
```

```python
swimming = olympics.loc[olympics['Sport'] == 'Swimming'].copy()
```

```python
swimming
```

```python
# fancy grouping magic
result = swimming.groupby(["NOC", "ID"], sort=True).apply(lambda x: x.sort_values("Year", ascending = False)).reset_index(drop=True)
```

```python
result.shape
```

```python
# drop the columns that we don't need
result = result.drop(columns=['Height', 'Weight', 'Team'])
result
```

```python
# get two dataframes: one with all of the NaNs, the other with all NaNs dropped
null = result[result['Medal'].isnull()]
null_dropped = result.dropna()

# check if the ones with NaN are in the non-NaN dataframe 
# - if so, we keep it in the origining dataframe
# - otherwise, we drop it from the origining dataframe
for row in null.itertuples():
    id_num = row[1]
    is_in = (null_dropped['ID'] == id_num).any()
    if not is_in:
        result = result[result.ID != id_num]
```

```python
result = result.replace(to_replace=["FRG, GDR, EUA"], value="GER")

checking = result.loc[result['NOC'] == 'GER']

```
