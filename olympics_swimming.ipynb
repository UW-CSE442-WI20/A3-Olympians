```python
import pandas as pd
import numpy as np
import re

regex = re.compile('[^a-zA-Z ]')
```

```python
olympics = pd.read_csv('../athlete_events.csv')
```

```python
olympics.head()
```

```python
swimming = olympics.loc[olympics['Sport'] == 'Swimming'].copy()

swimming
```

```python
#swimming_copy = swimming.copy()
swimming['Name'] = swimming['Name'].apply(lambda x: re.sub(regex, '', str(x)))

swimming['Last_Name'] = swimming['Name'].apply(lambda x:pd.Series(x.split()[::-1]))[0]
swimming
```

```python
swimming = swimming.sort_values(by=['Last_Name'], ascending=True)

swimming['Order'] = pd.factorize(swimming['Last_Name'])[0]

swimming

# swimming['Order'] = np.arange(len(swimming_copy))
# swimming
```

```python
# fancy grouping magic
swimming = swimming.replace(to_replace=["FRG", "GDR", "EUA"], value="GER")
result = swimming.groupby(["NOC", "ID"], sort=True).apply(lambda x: x.sort_values("Year", ascending = False)).reset_index(drop=True)
```

```python
result.shape
```

```python
# drop the columns that we don't need
result = result.drop(columns=['Height', 'Weight', 'Team'])
result
```

```python
# get two dataframes: one with all of the NaNs, the other with all NaNs dropped
null = result[result['Medal'].isnull()]
null_dropped = result.dropna()

# check if the ones with NaN are in the non-NaN dataframe 
# - if so, we keep it in the origining dataframe
# - otherwise, we drop it from the origining dataframe
for row in null.itertuples():
    id_num = row[1]
    is_in = (null_dropped['ID'] == id_num).any()
    if not is_in:
        result = result[result.ID != id_num]
```

```python
start_end = pd.read_csv('olympic_start_end_years.csv')
start_end
```

```python
result["Start"] = " "
result["End"] = " "
result['ID'] = result['ID'].astype('Int64')
result['Age'] = result['Age'].astype('Int64')
result['Sex'].replace('M', 0, inplace=True)
result['Sex'].replace('F', 1, inplace=True)
for idx, id_n, firstyear, lastyear in start_end.itertuples():
    result.loc[result['ID'] == id_n, 'Start'] = firstyear
    result.loc[result['ID'] == id_n, 'End'] = lastyear
    
result
```

```python
export_csv = result.to_csv (r'export_dataframe.csv', index = None, header=True)
```
